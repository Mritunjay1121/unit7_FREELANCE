```python
merged_data['Unemployment Rate (%)'] = pd.to_numeric(merged_data['Unemployment Rate (%)'], errors='coerce')
merged_data['% Under US $5.50 Per Day'] = pd.to_numeric(merged_data['% Under US $5.50 Per Day'], errors='coerce')
```

```python


# Define the independent variable (X) and the dependent variable (y)
X = merged_data['Unemployment Rate (%)']
y = merged_data['% Under US $5.50 Per Day']

# Add a constant term to the independent variable (intercept)
X = sm.add_constant(X)


```

### Report with Charts


```python
# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X['Unemployment Rate (%)'], y, label='Data Points', color='blue')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Poverty Rate (%)')

# Regression Line Plot
plt.plot(X['Unemployment Rate (%)'], model.predict(X), label='Regression Line', color='red')
plt.legend()
plt.title('Relationship between Poverty Rate and Unemployment Rate')
plt.grid(True)

# Show the plots
plt.show()
```
